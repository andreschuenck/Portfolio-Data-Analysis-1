13) Sabendo que A = {0, 1, 2, 3, 4, 5, 6}, B = {6, 7, 8, 9} e C = {2, 4, 6, 8, 10}, quais são os elementos do conjunto (A∩B)UC?

```python
A = {0, 1, 2, 3, 4, 5, 6}
B = {6, 7, 8, 9}
C = {2, 4, 6, 8, 10}
```

```python
def intersec(listaA, listaB):
    listaAB = []
    for x in listaA:
        if x in listaB:
            listaAB.append(x)
    return listaAB

listaAB = intersec(A, B)

def uni(listaB, listaC):
    listaAC = []
    for x in listaB:
        listaAC.append(x)
    for x in listaC:
        listaAC.append(x)
    return listaAC

print(uni(listaAB, C))
```
